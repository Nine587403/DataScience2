```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```
x = np.arange(0,20)
y = x**2
plt.plot(x,y)
plt.title('First Time With Matplotlib')
plt.xlabel('x')
plt.ylabel('y')
```

```
x = np.arange(0,100)
y = np.sin(x)
plt.plot(x,y, c = 'm')
plt.title('Sin Graph')
plt.xlabel('x')
plt.ylabel('y')
```

```
y = np.arange(0,100)
x = y**2
plt.plot(x,y, c = '#CCFF66')
plt.title('Graph')
plt.xlabel('x')
plt.ylabel('y')
```

```
x = np.arange(0,100)
y = np.cos(x)
plt.plot(x,y, c = 'r', ls = '--')
plt.title('Graph')
plt.xlabel('x')
plt.xlabel('y')
```

```
x = np.arange(0,20)
y1 = x**2
y2 = x**3

plt.plot(x,y1, c = 'r', marker = 'o')
plt.plot(x,y2, c = 'b', marker = '^')
plt.title('Graph')
plt.xlabel('x')
plt.xlabel('y')
```

```
x = np.arange(0,20)
y = x**2
x2 = y**2

plt.plot(x,y, c = 'r', marker = 'o', label = 'y = x**2 line')
plt.plot(x2,y, c = 'b', marker = '^', label = 'x = y**2 line')
plt.title('Graph')
plt.legend(loc=0)
plt.xlabel('x')
plt.xlabel('y')
```

```
y = np.arange(0,100)
x = y**2+4*y
plt.plot(x,y, c = '#FFFF00')
plt.title('Graph')
plt.xlabel('x')
plt.ylabel('y')
```

```
y = np.arange(0,100)
x = y**3
y2 = x**2+2
fig = plt.figure(figsize=[7,5])
ax1 = fig.add_axes([0.1,0.1,1,1])
ax1.plot(x,y)
ax2 = fig.add_axes([0.65,0.15,0.4,0.4])
ax2.plot(x,y2)
```

```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('https://github.com/Nine587403/DataScience2/blob/main/train.csv?raw=true')
Group_Class = df.groupby(['Pclass']).count()['PassengerId']
plt.figure(figsize=[5,5])
labelsClass = ['1st class', '2nd class', '3rd class' ]
plt.pie(Group_Class,labels = labelsClass)
plt.show()
```

```
Group_Embark = df.groupby(['Embarked']).count()['PassengerId']
plt.figure(figsize = [5,5])
exEmbark = [0,0.1,0]
labelsEmbark = ['Cherbourg', 'Queenstown', 'Southampton']
plt.pie(Group_Embark, labels = labelsEmbark, explode = exEmbark)
plt.show()
```

```
Group_Fare = df.groupby(['Fare']).count()['PassengerId']
plt.figure(figsize = [8,5])
plt.hist(Group_Fare, bins = 30)
plt.show()
```

```
df10 = pd.read_csv('https://github.com/Nine587403/DataScience2/blob/main/train.csv?raw=true')
df10['LastName'] = df10['Name'].str.split(',', expand=True)[0]
df11 = df10[df10['LastName'].duplicated()]
df12 = df11.groupby(['LastName']).size()

plt.figure(figsize = [50,5])
plt.bar(df12.index, df12.values)
plt.xticks(rotation=90)
plt.show()
```
