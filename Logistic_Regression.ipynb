```
avg_age = df['age'].mean()
df['age'].fillna(value=avg_age, inplace=True)
```

```
avg_bp = df['bp'].mean()
df['bp'].fillna(value=avg_bp, inplace=True)
```

```
avg_sg = df['sg'].mean()
df['sg'].fillna(value=avg_sg, inplace=True)
```

```
df.dropna(subset=['al', 'su', 'rbc'],inplace=True)
```

```
rbc = pd.get_dummies(df['rbc'], drop_first=True)
df = pd.concat([df, rbc], axis=1)
df.rename(columns={'normal':'rbc_normal'}, inplace=True)
```

```
df.dropna(subset=['pc'],inplace=True)
```

```
pc = pd.get_dummies(df['pc'], drop_first=True)
df = pd.concat([df, pc], axis=1)
df.rename(columns={'normal':'pc_normal'}, inplace=True)
```

```
df.dropna(subset=['pcc'],inplace=True)
```

```
pcc = pd.get_dummies(df['pcc'], drop_first=True)
df = pd.concat([df, pcc], axis=1)
df.rename(columns={'present':'pcc_present'}, inplace=True)
```

```
df.dropna(subset=['ba'],inplace=True)
```

```
ba = pd.get_dummies(df['ba'], drop_first=True)
df = pd.concat([df, ba], axis=1)
df.rename(columns={'present':'ba_present'}, inplace=True)
```

```
df.drop(['rbc', 'pc', 'pcc', 'ba'], 1, inplace=True)
```

```
avg_bgr = df['bgr'].mean()
df['bgr'].fillna(value=avg_bgr, inplace=True)
```

```
avg_bu = df['bu'].mean()
df['bu'].fillna(value=avg_bu, inplace=True)
```

```
avg_sc = df['sc'].mean()
df['sc'].fillna(value=avg_sc, inplace=True)
```

```
avg_sod = df['sod'].mean()
df['sod'].fillna(value=avg_sod, inplace=True)
```

```
avg_pot = df['pot'].mean()
df['pot'].fillna(value=avg_pot, inplace=True)
```

```
avg_hemo = df['hemo'].mean()
df['hemo'].fillna(value=avg_hemo, inplace=True)
```

```
df['pcv'].replace('\t?', np.NaN ,inplace=True)
df['pcv'] = df['pcv'].astype('float')
```

```
avg_pcv = df['pcv'].mean()
df['pcv'].fillna(value=avg_pcv, inplace=True)
```

```
df['wc'] = df['wc'].astype('float')
avg_wc = df['wc'].mean()
df['wc'].fillna(value=avg_wc, inplace=True)
```

```
df['rc'] = df['rc'].astype('float')
avg_rc = df['rc'].mean()
df['rc'].fillna(value=avg_rc, inplace=True)
```

```
df.dropna(subset=['htn'],inplace=True)
```

```
htn = pd.get_dummies(df['htn'], drop_first=True)
df = pd.concat([df, htn], axis=1)
df.rename(columns={'yes':'htn_yes'}, inplace=True)
```

```
dm = pd.get_dummies(df['dm'], drop_first=True)
df = pd.concat([df, dm], axis=1)
df.rename(columns={'yes':'dm_yes'}, inplace=True)
```

```
df['cad'].replace('\tno', 'no',inplace=True)
```

```
cad = pd.get_dummies(df['cad'], drop_first=True)
df = pd.concat([df, cad], axis=1)
df.rename(columns={'yes':'cad_yes'}, inplace=True)
```

```
df.dropna(subset=['appet'],inplace=True)
```

```
appet = pd.get_dummies(df['appet'], drop_first=True)
df = pd.concat([df, appet], axis=1)
df.rename(columns={'poor':'appet_poor'}, inplace=True)
```

```
pe = pd.get_dummies(df['pe'], drop_first=True)
df = pd.concat([df, pe], axis=1)
df.rename(columns={'yes':'pe_yes'}, inplace=True)
```

```
ane = pd.get_dummies(df['ane'], drop_first=True)
df = pd.concat([df, ane], axis=1)
df.rename(columns={'yes':'ane_yes'}, inplace=True)
```

```
df['classification'].replace('ckd\t', 'ckd',inplace=True)
```

```
calssification = pd.get_dummies(df['classification'], drop_first=True)
df = pd.concat([df, calssification], axis=1)
df.rename(columns={'notckd':'class_notckd'}, inplace=True)
```

```
df.drop(['htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'classification'], 1, inplace = True)
```

```
df.info()
```

```
fig = plt.figure(figsize=(30,17))
sns.heatmap(df.corr(), annot=True)
```

```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score 
from sklearn.metrics import confusion_matrix
```

```
X = df[['bu', 'sc']]
y = df['class_notckd']
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```
logistic_regression = LogisticRegression()
logistic_regression.fit(X_train, y_train)
```

```
predicted = logistic_regression.predict(X_test)
```

```
print(confusion_matrix(y_test,predicted))
```

```
print('Accuracy = ', accuracy_score(y_test, predicted))
print('F1-Score = ', f1_score(y_test, predicted))
print('Precision = ', precision_score(y_test, predicted))
print('Recall = ', recall_score(y_test, predicted))
```
