```
from bs4 import BeautifulSoup
import requests

url = 'https://www.goldtraders.or.th/'
response = requests.get(url)
```

```
c = response.content
soup  = BeautifulSoup(c,'html.parser')
print(soup.prettify())
```

```
all = soup.find(attrs={'id':'DetailPlace_uc_goldprices1_GoldPricesUpdatePanel'})
```

```
gold = []
bs = []
price = []

for i in range(1,len(all.findAll('tr'))):
  if i%2 != 0:
    gold.append((all.findAll('tr')[i].text.split()[0]+' '+all.findAll('tr')[i].text.split()[1]))
    bs.append(all.findAll('tr')[i].text.split()[2])
    price.append(all.findAll('tr')[i].text.split()[3])
  else:
    gold.append((all.findAll('tr')[i-1].text.split()[0]+' '+all.findAll('tr')[i-1].text.split()[1]))
    bs.append(all.findAll('tr')[i].text.split()[0])
    price.append(all.findAll('tr')[i].text.split()[1])
```

```
gold ,bs ,price
```

```
import pandas as pd
```

```
df = pd.DataFrame()
```

```
df['gold'] = gold
df['bs'] = bs
df['price'] = price

df
```
