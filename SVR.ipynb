```
from sklearn.model_selection import train_test_split
```

```
X = df['sqft_living']
y = df['price']
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```
X_train = np.array(X_train).reshape(-1,1)
X_test = np.array(X_test).reshape(-1,1)
y_train = np.array(y_train).reshape(-1,1)
y_test = np.array(y_test).reshape(-1,1)
```

```
from sklearn.preprocessing import StandardScaler
sc_X_train = StandardScaler()
sc_y_train = StandardScaler()

X_train = sc_X_train.fit_transform(X_train)
y_train = sc_y_train.fit_transform(y_train)
```

```
from sklearn.svm import SVR
regressor = SVR(kernel = 'linear')
regressor.fit(X_train, y_train)
```

```
result_linear = regressor.predict(sc_X_train.transform(X_test))
```

```
result_linear = sc_y_train.inverse_transform(result_linear)
```

```
fig = plt.figure(figsize=(14,9))
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, result_linear, color='blue')
plt.title('SVR')
plt.xlabel('sqft_living')
plt.ylabel('Price')
plt.show()
```

```
from sklearn import metrics
```

```
print('MAE', metrics.mean_absolute_error(result_linear, y_test))
print('MSE', metrics.mean_squared_error(result_linear, y_test))
print('RMSE', np.sqrt(metrics.mean_squared_error(result_linear, y_test)))
```

```
regressor2 = SVR(kernel = 'rbf')
regressor2.fit(X_train, y_train)
```

```
X2_test = X_test.tolist()
X2_test.sort()
X3_test = np.asarray(X2_test)
```

```
result_rbf = regressor2.predict(sc_X_train.transform(X3_test))
```

```
result_rbf = sc_y_train.inverse_transform(result_rbf)
```

```
fig = plt.figure(figsize=(12,8))
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X3_test, result_rbf, color='blue')
plt.title('SVR')
plt.xlabel('sqft_living')
plt.ylabel('Price')
plt.show()
```

```
result_rbf2 = regressor2.predict(sc_X_train.transform(X_test))
result_rbf2 = sc_y_train.inverse_transform(result_rbf2)
```

```
print('MAE', metrics.mean_absolute_error(result_rbf2, y_test))
print('MSE', metrics.mean_squared_error(result_rbf2, y_test))
print('RMSE', np.sqrt(metrics.mean_squared_error(result_rbf2, y_test)))
```

```
X_mul_train, X_mul_test, y_mul_train, y_mul_test = train_test_split(X_mul, y_mul, test_size = 0.2, random_state = 100)
```
